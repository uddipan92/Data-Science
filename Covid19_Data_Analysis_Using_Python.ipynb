### Task 2.1: importing covid19 dataset

### Task 2.2: Delete the useless columns

```python
covid19_df.drop(["Lat", "Long"], axis=1, inplace=True)
```

### Task 2.3: Aggregating the rows by the country

```python
covid19_df_aggregated=covid19_df.groupby("Country/Region").sum()
```

### Task 2.4: Visualizing data related to a country for example China

### Task3: Calculating a good measure

### task 3.1: caculating the first derivative of the curve

```python
covid19_df_aggregated.loc["China"].diff().plot()
```

### task 3.2: find maxmimum infection rate for China

```python
covid19_df_aggregated.loc["China"].diff().max()
```

```python
#find maxmimum infection rate for Italy
covid19_df_aggregated.loc["Italy"].diff().max()
```

```python
#find maxmimum infection rate for Spain
covid19_df_aggregated.loc["Spain"].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries=list(covid19_df_aggregated.index)
max_infection_rates=[]
for i in countries:
    max_infection_rates.append(covid19_df_aggregated.loc[i].diff().max())

covid19_df_aggregated["max_infection_rates"]=max_infection_rates
```

```python
covid19_df_aggregated.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
covid19_newdf=pd.DataFrame(covid19_df_aggregated["max_infection_rates"])
```

```python
covid19_newdf.head()
```
