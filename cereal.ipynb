```python
print(df_cereal['mfr'].unique())
```

```python
manufacturer_mapping = {'N': 'Nabisco',
                        'Q': 'Quaker Oats',
                        'K': 'Kelloggs',
                        'R': 'Ralston Purina',
                        'G': 'General Mills',
                        'P': 'Post',
                        'A': 'American Home Foods Products'}
```

```python
df_cereal['manufacturer'] = df_cereal['mfr'].map(manufacturer_mapping)
```

```python
df_cereal.head()
```

```python
import seaborn as sns
```

```python
sns.set(style="darkgrid")

mfr_cereal_counts = df_cereal['manufacturer'].value_counts()

sns.barplot(x=mfr_cereal_counts.values, y=mfr_cereal_counts.index)

plt.title("Number of cereals manufactured by the manufacturer")
plt.xlabel("Count")
plt.ylabel("Manufacturer")

plt.show()
```

```python
y=df_cereal['rating']
```

```python
numerical_predators=df_cereal.select_dtypes(include='number')
print(numerical_predators)
```

```python
numerical_predators.head()
```

```python
x=numerical_predators.iloc[:,:-1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
#fitting our model to train and test
lm = LinearRegression()
model = lm.fit(x_train,y_train)
```

```python
print("X_train shape:", x_train.shape)
print("X_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
pred_y=lm.predict(x_test)
```

```python
plt.scatter(y_test,pred_y,color='red')
```

```python
from sklearn.metrics import mean_squared_error
print("Mean Squared error on the test dataset is: ",mean_squared_error(y_test,pred_y))
```
